### Import Amplitude Encoding Feature Map

```python
# If you are not using the `qdna-lib` source code,
# comment out the following two lines.
import sys
sys.path.append('../../qdna-lib/')

from qdna.embedding import AeFeatureMap
```

### Create an embedding.

```python
from qiskit import QuantumCircuit

num_qubits = 4

embedding = AeFeatureMap(num_qubits, reps=1)

circuit = QuantumCircuit(num_qubits)
circuit.compose(embedding, inplace=True, wrap=True)

circuit.draw('mpl', style="iqp")
```
